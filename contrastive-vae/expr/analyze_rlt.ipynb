```python
import json
import os
import pandas as pd
```

```python
K = 5
cnn_acc_list, cnn_roc_list, cnn_pr_list = [], [], []
cdvae_acc_list, cdvae_roc_list, cdvae_pr_list = [], [], []

for file_name in os.listdir("../expr_output/cmnist/cls"):
    if file_name.startswith(f"cmnist-k{K}") and file_name.endswith(".json"):
        with open(f"../expr_output/cmnist/cls/{file_name}", "r") as f:
            data = json.load(f)
            cnn_acc_list.append(data["cnn"]["acc"])
            cnn_pr_list.append(data["cnn"]["pr"]["overall"])
            cnn_roc_list.append(data["cnn"]["roc"]["overall"])

            cdvae_acc_list.append(data["vae + mlp"]["acc"])
            cdvae_pr_list.append(data["vae + mlp"]["pr"]["overall"])
            cdvae_roc_list.append(data["vae + mlp"]["roc"]["overall"])
df = pd.DataFrame({
    'cnn_acc': cnn_acc_list,
    'cdvae_acc': cdvae_acc_list,
    'cnn_roc': cnn_roc_list,
    'cdvae_roc': cdvae_roc_list,
    'cnn_pr': cnn_pr_list,
    'cdvae_pr': cdvae_pr_list,
})
df.describe().iloc[3:, :].loc[["50%", "25%", "75%", "min", "max"],:]
```

```python
for K in range(1, 6):
    rel_acc_list, rel_roc_list, rel_pr_list = [], [], []

    for file_name in os.listdir("../expr_output/cmnist/cls"):
        if file_name.startswith(f"cmnist-k{K}") and file_name.endswith(".json"):
            with open(f"../expr_output/cmnist/cls/{file_name}", "r") as f:
                data = json.load(f)
                rel_acc_list.append(
                    data["vae + mlp"]["acc"] / data["cnn"]["acc"])
                rel_pr_list.append(
                    data["vae + mlp"]["pr"]["overall"] / data["cnn"]["pr"]["overall"])
                rel_roc_list.append(
                    data["vae + mlp"]["roc"]["overall"] / data["cnn"]["roc"]["overall"])

    df_rel = pd.DataFrame({
        'rel_acc': rel_acc_list,
        'rel_roc': rel_roc_list,
        'rel_pr': rel_pr_list
    })
    print(df_rel.describe().iloc[3:, :].loc[["50%", "25%", "75%", "min", "max"],:])
```
