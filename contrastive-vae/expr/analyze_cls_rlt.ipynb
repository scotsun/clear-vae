```python
import json
import os
import pandas as pd
```

```python
K = 5
cnn_acc_list, cnn_roc_list, cnn_pr_list = [], [], []
mlvae_acc_list, mlvae_roc_list, mlvae_pr_list = [], [], []
clearvae_acc_list, clearvae_roc_list, clearvae_pr_list = [], [], []

for file_name in os.listdir("../expr_output/cmnist/classification"):
    if file_name.startswith(f"cmnist-k{K}") and file_name.endswith(".json"):
        with open(f"../expr_output/cmnist/classification/{file_name}", "r") as f:
            data = json.load(f)
            cnn_acc_list.append(data["cnn"]["acc"])
            cnn_pr_list.append(data["cnn"]["pr"]["overall"])
            cnn_roc_list.append(data["cnn"]["roc"]["overall"])

            mlvae_acc_list.append(data["ml-vae + mlp"]["acc"])
            mlvae_pr_list.append(data["ml-vae + mlp"]["pr"]["overall"])
            mlvae_roc_list.append(data["ml-vae + mlp"]["roc"]["overall"])

            clearvae_acc_list.append(data["clear-vae + mlp"]["acc"])
            clearvae_pr_list.append(data["clear-vae + mlp"]["pr"]["overall"])
            clearvae_roc_list.append(data["clear-vae + mlp"]["roc"]["overall"])
df = pd.DataFrame({
    'cnn_acc': cnn_acc_list,
    'mlvae_acc': mlvae_acc_list,
    'clearvae_acc': clearvae_acc_list,
    'cnn_roc': cnn_roc_list,
    'mlvae_roc': mlvae_roc_list,
    'clearvae_roc': clearvae_roc_list,
    'cnn_pr': cnn_pr_list,
    'mlvae_pr': mlvae_pr_list,
    'clearvae_pr': clearvae_pr_list
})
df.describe().iloc[3:, :].loc[["50%", "25%", "75%", "min", "max"],:]
```

```python
for K in range(1, 6):
    rel_acc_list, rel_roc_list, rel_pr_list = [], [], []

    for file_name in os.listdir("../expr_output/cmnist/classification"):
        if file_name.startswith(f"cmnist-k{K}") and file_name.endswith(".json"):
            with open(f"../expr_output/cmnist/classification/{file_name}", "r") as f:
                data = json.load(f)
                rel_acc_list.append(
                    data["clear-vae + mlp"]["acc"] / data["cnn"]["acc"])
                rel_pr_list.append(
                    data["clear-vae + mlp"]["pr"]["overall"] / data["cnn"]["pr"]["overall"])
                rel_roc_list.append(
                    data["clear-vae + mlp"]["roc"]["overall"] / data["cnn"]["roc"]["overall"])

    df_rel = pd.DataFrame({
        'rel_acc': rel_acc_list,
        'rel_roc': rel_roc_list,
        'rel_pr': rel_pr_list
    })
    print(df_rel.describe().iloc[3:, :].loc[["50%", "25%", "75%", "min", "max"],:].transpose().round(3))
```
